```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
df = pd.read_csv('../data/processed/travel_advisories.csv')

# Display basic information
print(df.info())
print(df.describe())

# Display the first few rows
df.head()
```

```python
# Plot the distribution of risk scores
plt.figure(figsize=(10, 6))
sns.histplot(df['Risk Score'], bins=20, kde=True)
plt.title('Distribution of Travel Risk Scores')
plt.xlabel('Risk Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# High-risk countries
high_risk = df[df['Risk Score'] > 4]
print("High-Risk Countries:")
print(high_risk[['Country', 'Risk Score']])

# Low-risk countries
low_risk = df[df['Risk Score'] < 2]
print("Low-Risk Countries:")
print(low_risk[['Country', 'Risk Score']])
```

```python
# Display advisory messages for high-risk countries
print("Advisory Messages for High-Risk Countries:")
print(high_risk[['Country', 'Advisory Message']].to_string(index=False))

# Display advisory messages for low-risk countries
print("Advisory Messages for Low-Risk Countries:")
print(low_risk[['Country', 'Advisory Message']].to_string(index=False))
```
